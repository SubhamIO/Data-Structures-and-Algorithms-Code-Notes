## Level Order Linewise  - Zig Zag Printing

```python
class Node():
 
    def __init__(self, data):
        self.data = data
        self.children = []
def construct(arr):
    stack = []
    for i in range(len(arr)):
        if arr[i] == -1:
            stack.pop()
        else:
            # create Node with the data
            t = Node(arr[i])

            #If stack already has an element
            if len(stack)>0:
                #stack[-1] would give the topmost Node
                #add the new Node as its children
                stack[-1].children.append(t) 

                #Push it to stack
                stack.append(t)
            #If the stack is empty then declare the first Node as root
            else:
                root = t
                #push it to stack
                stack.append(root)
    return root

            
def levelOrderLinewiseZigZag(node):
    #using 2 queues
    ms = [] #main queue
    cs = [] #child queue
    #First add the Root node to Main Queue
    ms.append(node)
    level = 1 #make this 0 to get opposite zip
    while(len(ms) > 0):
        #Printing top to bottom of stack
        node = ms.pop() #pop last element
        print(str(node.data),end = " ")
        
        #The direction in which you enter children to child stack,
        #They will be later printed in opposite direction
        if level%2 != 0:
            for i in range (len(node.children)):
                cs.append(node.children[i])
        else:
            for i in range(len(node.children)-1,-1,-1):
                cs.append(node.children[i])
            
        #New Level   
        if len(ms) == 0:
            level +=1
            ms = cs
            cs = []
            print()  
    
if __name__ == '__main__':
    arr = [10,20,50,-1,60,-1,-1,30,70,-1,80,110,-1,120,-1,-1,90,-1,-1,40,100,-1,-1,-1]
    root = construct(arr)
    levelOrderLinewiseZigZag(root)
  
```

## Remove Leaves from a Generic tree

```python
class Node():
 
    def __init__(self, data):
        self.data = data
        self.children = []
def construct(arr):
    stack = []
    for i in range(len(arr)):
        if arr[i] == -1:
            stack.pop()
        else:
            # create Node with the data
            t = Node(arr[i])

            #If stack already has an element
            if len(stack)>0:
                #stack[-1] would give the topmost Node
                #add the new Node as its children
                stack[-1].children.append(t) 

                #Push it to stack
                stack.append(t)
            #If the stack is empty then declare the first Node as root
            else:
                root = t
                #push it to stack
                stack.append(root)
    return root
'''
- EXP : display(10) will print itself and its family
- FAITH : display(20),display(30),display(40) will print themselves and their families
- EXP-FAITH : display(10) = self(10)+display(20)+display(30)+display(40)'''
#If I send Node(10) as input, it will print
def display(node):
    string = str(node.data) + "-->"
    for child in node.children:
        string+= str(child.data) + ", " #prints its family, whose parent has been passed
        
    string+="."
    print(string)
    
    for child in node.children: # print all the families
        display(child)
                    
                
def removeLeaves(node):
    
    '''Removal Logic should be written in PreOrder(before recursion) fashion
    else it would remove the node which was not leaf node, because after removal of its leaves,
    it would also become a leaf, and if it was being checked as Potorder it would create this problem'''
    for child in node.children[::-1]:#Removal should be from last to first, else after removal from first it would shift index
        if len(child.children) == 0:#If child node dosnt have children, then it is leaf node
            node.children.remove(child) #so, remove that child from list
            
    #Faith : each would remove their leaf nodes
    for child in node.children:
        removeLeaves(child)
        
    
if __name__ == '__main__':
    arr = [10,20,50,-1,60,-1,-1,30,70,-1,80,110,-1,120,-1,-1,90,-1,-1,40,100,-1,-1,-1]
    root = construct(arr)
    removeLeaves(root)
    display(root)
```

## Linearize a Generic Tree in Preorder fashion

__Unoptimised Approach = O(n^2)__

```python
class Node():
 
    def __init__(self, data):
        self.data = data
        self.children = []
def construct(arr):
    stack = []
    for i in range(len(arr)):
        if arr[i] == -1:
            stack.pop()
        else:
            # create Node with the data
            t = Node(arr[i])

            #If stack already has an element
            if len(stack)>0:
                #stack[-1] would give the topmost Node
                #add the new Node as its children
                stack[-1].children.append(t) 

                #Push it to stack
                stack.append(t)
            #If the stack is empty then declare the first Node as root
            else:
                root = t
                #push it to stack
                stack.append(root)
    return root
'''
- EXP : display(10) will print itself and its family
- FAITH : display(20),display(30),display(40) will print themselves and their families
- EXP-FAITH : display(10) = self(10)+display(20)+display(30)+display(40)'''
#If I send Node(10) as input, it will print
def display(node):
    string = str(node.data) + "-->"
    for child in node.children:
        string+= str(child.data) + ", " #prints its family, whose parent has been passed
        
    string+="."
    print(string)
    
    for child in node.children: # print all the families
        display(child)
                

def getTail(node):
    while(len(node.children) == 1):
        node = node.children[0]
    return node
    
def linearize(node):
    for child in  node.children:
        linearize(child)
        
    while (len(node.children) >1):
        
        last_child_node = node.children.pop(len(node.children)-1)
        #Fetch second_last after removing last element, so second_last now become last
        second_last_child_node = node.children[len(node.children)-1] 
        second_last_tail = getTail(second_last_child_node)
        second_last_tail.children.append(last_child_node)

           
    
if __name__ == '__main__':
    arr = [10,20,50,-1,60,-1,-1,30,70,-1,80,110,-1,120,-1,-1,90,-1,-1,40,100,-1,-1,-1]
    root = construct(arr)
    linearize(root)
    display(root)
```

```python
class Node():
 
    def __init__(self, data):
        self.data = data
        self.children = []
def construct(arr):
    stack = []
    for i in range(len(arr)):
        if arr[i] == -1:
            stack.pop()
        else:
            # create Node with the data
            t = Node(arr[i])

            #If stack already has an element
            if len(stack)>0:
                #stack[-1] would give the topmost Node
                #add the new Node as its children
                stack[-1].children.append(t) 

                #Push it to stack
                stack.append(t)
            #If the stack is empty then declare the first Node as root
            else:
                root = t
                #push it to stack
                stack.append(root)
    return root



def multisolver(node,depth):
    #These global parameters are in heap , they dont depend on Euler path
    #The parameter like depth is in Stack, they change with Euler path(incr or decr wrt Euler path)
    global size
    global min_val
    global max_val
    global height
    
    size+=1
    min_val = min(min_val,node.data)
    max_val = max(max_val,node.data)
    height = max(height,depth) #Height = depth of deepest node
    #This code is to traverse the entire tree
    for child in node.children:
        multisolver(child,depth+1) #when u go froma node to its child depth increases by 1

    
if __name__ == '__main__':
    arr = [10,20,50,-1,60,-1,-1,30,70,-1,80,110,-1,120,-1,-1,90,-1,-1,40,100,-1,-1,-1]
    root = construct(arr)
    
    #Assign values here and in multisolver mention them as global var
    size = 0
    min_val = float('inf')
    max_val = float('-inf')
    height = 0
    
    multisolver(root,0)
    print('Size: ', size)
    print('Min Value: ',min_val)
    print('Max Value: ',max_val)
    print('Height: ',height)
```

```python
class Node():
 
    def __init__(self, data):
        self.data = data
        self.children = []
def construct(arr):
    stack = []
    for i in range(len(arr)):
        if arr[i] == -1:
            stack.pop()
        else:
            # create Node with the data
            t = Node(arr[i])

            #If stack already has an element
            if len(stack)>0:
                #stack[-1] would give the topmost Node
                #add the new Node as its children
                stack[-1].children.append(t) 

                #Push it to stack
                stack.append(t)
            #If the stack is empty then declare the first Node as root
            else:
                root = t
                #push it to stack
                stack.append(root)
    return root



def predecessorAndSuccessor(node,data):
    #These global parameters are in heap , they dont depend on Euler path
    #The parameter like depth is in Stack, they change with Euler path(incr or decr wrt Euler path)
    global predecessor
    global successor
    global state #when we get the node make it 1 and next node to 2.So predecessor state= 0,successor =2 ,element=1
    
    #keep making the nodes as Predecessor
    #when we get the node make its state as 1, so predecessor will be the prev node with state 0
    if state == 0:
        if node.data == data:
            state = 1
        else:
            predecessor = node
    elif state == 1:
        successor = node
        state = 2
    
    #This code is to traverse the entire tree
    for child in node.children:
        predecessorAndSuccessor(child,data) #when u go froma node to its child depth increases by 1

    
if __name__ == '__main__':
    arr = [10,20,50,-1,60,-1,-1,30,70,-1,80,110,-1,120,-1,-1,90,-1,-1,40,100,-1,-1,-1]
    root = construct(arr)
    data = 90
    
    #Assign values here and in multisolver mention them as global var
    predecessor = None
    successor = None
    state = 0
    
    predecessorAndSuccessor(root,data)
    if predecessor == None:
        print('Predecessor = Not Found')
    else:
        print('Predecessor = {}'.format(predecessor.data))
        
    if successor == None:
        print('Successor = Not Found')
    else:
        print('Successor = {}'.format(successor.data))
   
```

## Node with Maximum Subtree Sum

```python
class Node():
 
    def __init__(self, data):
        self.data = data
        self.children = []
def construct(arr):
    stack = []
    for i in range(len(arr)):
        if arr[i] == -1:
            stack.pop()
        else:
            # create Node with the data
            t = Node(arr[i])

            #If stack already has an element
            if len(stack)>0:
                #stack[-1] would give the topmost Node
                #add the new Node as its children
                stack[-1].children.append(t) 

                #Push it to stack
                stack.append(t)
            #If the stack is empty then declare the first Node as root
            else:
                root = t
                #push it to stack
                stack.append(root)
    return root

'''Follows:
return smething else and calculate something else strategy
- Here we are calculating sum which we wont return, while in the journey of calculating sum
we calculate the Node with max sum subtree'''
def maxSumSubtree(node):   
    #Here we will get the max node from the childs
    sum=0
    global max_sum_node
    global max_sum
    
    
    for child in node.children:
        child_sum = maxSumSubtree(child)
        sum+=child_sum #sum of childrens
    sum+=node.data   #sum of childrens and itself
    
    if sum > max_sum:
        max_sum_node = node.data
        max_sum = sum
                
    return sum       
    
if __name__ == '__main__':
    arr = [10,20,-50,-1,-60,-1,-1,30,-70,-1,80,-110,-1,120,-1,-1,90,-1,-1,40,-100,-1,-1,-1]
    root = construct(arr)
    #Assign values here and in multisolver mention them as global var
    max_sum_node = None
    max_sum = float('-inf')
    maxSumSubtree(root)
    print('MaxSum subtree Node is {} and Max sum is {} '.format(max_sum_node,max_sum))
  
```

## Diameter of Generic Tree

```python
class Node():
 
    def __init__(self, data):
        self.data = data
        self.children = []
def construct(arr):
    stack = []
    for i in range(len(arr)):
        if arr[i] == -1:
            stack.pop()
        else:
            # create Node with the data
            t = Node(arr[i])

            #If stack already has an element
            if len(stack)>0:
                #stack[-1] would give the topmost Node
                #add the new Node as its children
                stack[-1].children.append(t) 

                #Push it to stack
                stack.append(t)
            #If the stack is empty then declare the first Node as root
            else:
                root = t
                #push it to stack
                stack.append(root)
    return root

'''Follows:
return smething else and calculate something else strategy
- Here we are calculating diameter which we wont return, while in the journey of calculating height
we calculate the Diameter as well'''
def diameterOfGenericTree(node):   
    global diameter
    
    '''Find deepest child height and 2nd deepest child heights'''
    deepest_ht = -1
    second_deepest_ht = -1
    for child in node.children:
        child_ht = diameterOfGenericTree(child)
        
        if child_ht > deepest_ht:
            second_deepest_ht = deepest_ht
            deepest_ht = child_ht
            
        elif child_ht > second_deepest_ht:
            second_deepest_ht = child_ht
    
    calculated_diameter = deepest_ht + second_deepest_ht + 2
    diameter = max(diameter,calculated_diameter)
    
    deepest_ht+=1 # childs to node = 1 edge
    return deepest_ht       
    
if __name__ == '__main__':
    arr = [10,20,-50,-1,-60,-1,-1,30,-70,-1,80,-110,-1,120,-1,-1,90,-1,-1,40,-100,-1,-1,-1]
    root = construct(arr)
    #Assign values here and in multisolver mention them as global var
    diameter = 0
    diameterOfGenericTree(root)
    print('Diamter of Generic Tree is {} '.format(diameter))
  
```

## Iterative Preorder and Postorder of Generic Tree

```python
class Node():
 
    def __init__(self, data):
        self.data = data
        self.children = []
def construct(arr):
    stack = []
    for i in range(len(arr)):
        if arr[i] == -1:
            stack.pop()
        else:
            # create Node with the data
            t = Node(arr[i])

            #If stack already has an element
            if len(stack)>0:
                #stack[-1] would give the topmost Node
                #add the new Node as its children
                stack[-1].children.append(t) 

                #Push it to stack
                stack.append(t)
            #If the stack is empty then declare the first Node as root
            else:
                root = t
                #push it to stack
                stack.append(root)
    return root

class Pair(): 
    def __init__(self, node,state):
        self.node = node
        self.state = state       
def iterativePreorderAndPostorder(node):
    stack = []
    stack.append(Pair(node,-1)) #push node to stack, with state = -1
    
    pre = ""
    post = ""
    
    while len(stack) > 0:
        #pick the top element Pair
        top = stack[-1]
        
        if top.state == -1:
            pre+=str(top.node.data) + " "
            top.state+=1
            
        elif top.state == len(top.node.children):
            post+=str(top.node.data) + " "
            stack.pop()
        
        else:
            child_pair = Pair(top.node.children[top.state],-1)
            stack.append(child_pair)
            top.state+=1
    print('Preorder :',pre)
    print('Postorder :',post)        
        

if __name__ == '__main__':
    arr = [10,20,-1,30,50,-1,60,-1,-1,40,-1,-1]
    root = construct(arr)
    
    iterativePreorderAndPostorder(root)
  
```
