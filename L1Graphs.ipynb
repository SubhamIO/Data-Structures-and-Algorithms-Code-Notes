## Has Path ?

```python
class Edge():
    def __init__(self,src,nbr,wt):
        self.src = src #Source Vertex
        self.nbr = nbr #Neighbour Vertex
        self.wt = wt #Edge Weight 

def hasPath(graph,src,dst,visited):
    if src == dst:
        return True
    
    visited[src] = True
    #Now keep asking the neighours
    for edge in graph[src]: #Accessing list inside a list
        if visited[edge.nbr] == False: #if neighbor is not visited yet then call its neighbor
            hasNbrPath = hasPath(graph,edge.nbr,dst,visited) #path from its neighbour to destination
            if hasNbrPath == True:
                return True
        
    return False
if __name__ =='__main__':
    vertices = int(input())
    graph = [ [] for _ in range(vertices) ]
    
    edges = int(input())
    for i in range(edges):
        parts = input().split(" ")
        v1 = int(parts[0])
        v2 = int(parts[1])
        wt = int(parts[2])
    
        #Since it is biderectional so '0 1 10' and '1 0 10' both we need to store
        graph[v1].append(Edge(v1,v2,wt))
        graph[v2].append(Edge(v2,v1,wt))
    
    src = int(input())
    dst = int(input())
    
    visited = [False]*vertices
    path = hasPath(graph,src,dst,visited)
    print(path)
```

```python
class Edge():
    def __init__(self,src,nbr,wt):
        self.src = src #Source Vertex
        self.nbr = nbr #Neighbour Vertex
        self.wt = wt #Edge Weight 
        
class Pair():
    def __init__(self,wsf,psf):
        self.wsf = wsf
        self.psf = psf
    def __lt__(self,pair):
        return self.wsf < pair.wsf
        

spath = ""
spathwt = float('inf')
lpath = ""
lpathwt = float('-inf')
cpath = ""
cpathwt = float('inf')
fpath = ""
fpathwt = float('-inf')

import queue as Q
pq = Q.PriorityQueue()

def multisolver(graph,src,dst,visited,criteria,k,psf,wsf):
    global spath
    global lpath
    global cpath
    global fpath
    
    global spathwt
    global lpathwt
    global cpathwt
    global fpathwt
    
    global pq
    
    #psf = path so far
    #wsf = weight so far
    if src == dst:
        '''Smallest Path'''
        if wsf < spathwt:
            spathwt = wsf
            spath = psf
            
        '''Largest Path'''   
        if wsf > lpathwt:
            lpathwt = wsf
            lpath = psf
            
        '''Just Larger Path(Ceil)'''
        #greater than criteria , but smaller than those in ceil path
        #bado mein sabse chhota
        if wsf > criteria and wsf < cpathwt:
            cpathwt = wsf
            cpath = psf
            
        '''Just Smaller Path(Floor)'''   
        #lesser than criteria , but largest than those in floor path
        #chhoto mein sabse bada
        if wsf < criteria and wsf > fpathwt:
            fpathwt = wsf
            fpath = psf
            
        '''k-th Largest Path'''    
        if pq.qsize() < k:
            pq.put(Pair(wsf,psf))
        else:  
            #PriorityQueue : element present in ascending order
            #once u do pq.get() the element is removed from PriorityQueue
            least = pq.get() 
            #If u get an element greater than least among k, then add that
            if wsf > least.wsf:
                pq.put(Pair(wsf,psf))
            #else add the element which was removed back to PriorityQueue
            else:
                pq.put(least)
                
            
            
        return
    
    visited[src] = True
    #Now keep asking the neighours
    for edge in graph[src]: #Accessing list inside a list
        if visited[edge.nbr] == False: #if neighbor is not visited yet then call its neighbor
            multisolver(graph,edge.nbr,dst,visited,criteria,k,psf+str(edge.nbr),wsf+edge.wt) #path from its neighbour to destination
    visited[src] = False
    
if __name__ =='__main__':
    vertices = int(input())
    graph = [ [] for _ in range(vertices) ]
    
    edges = int(input())
    for i in range(edges):
        parts = input().split(" ")
        v1 = int(parts[0])
        v2 = int(parts[1])
        wt = int(parts[2])
    
        #Since it is biderectional so '0 1 10' and '1 0 10' both we need to store
        graph[v1].append(Edge(v1,v2,wt))
        graph[v2].append(Edge(v2,v1,wt))
    
    src = int(input())
    dst = int(input())
    
    criteria = int(input())
    k = int(input())
    
    visited = [False]*vertices
    multisolver(graph,src,dst,visited,criteria,k,str(src)+"",0) #send by ppending src from first, so its neighbours will attach next

    print('Smallest Path : ',spath+"@"+str(spathwt))
    print('Largest Path : ',lpath+"@"+str(lpathwt))
    print('Just Larger Path(Ceil) than {} = {}'.format(criteria,cpath+"@"+str(cpathwt)))
    print('Just Smaller Path(Floor) than {} = {}'.format(criteria,fpath+"@"+str(fpathwt)))
    kth_large = pq.get()
    print('{}th Largest Path = {}'.format(k,str(kth_large.psf)+"@"+str(kth_large.wsf)))
    
```

## Hamiltonian Path and Cycles in Graph

```python
class Edge():
    def __init__(self,src,nbr,wt):
        self.src = src #Source Vertex
        self.nbr = nbr #Neighbour Vertex
        self.wt = wt #Edge Weight 

def hamiltonian(graph,src,visited,path,original_src):
    #all are visited
    #from first visited was passed empty and path was appended with src,
    #so it contains 1 more, so visited will be 1 less
    if len(visited) == len(graph)-1:
        print(path,end = " ")
        
        cycle = False
        #Deciding whether Hamiltonian path or cycle
        for edge in graph[src]:#last source vertex
            if edge.nbr == original_src:#if it has any neighbor which is equal to original src then it is cycle
                cycle=True
                break
        if cycle:
            print("@Cycle")
        else:
            print("@Path")
        return
    
    visited.add(src)
    #Now keep asking the neighours
    for edge in graph[src]: #Accessing list inside a list
        if edge.nbr not in visited: #if neighbor is not visited yet then call its neighbor
            hamiltonian(graph,edge.nbr,visited,path+str(edge.nbr),original_src) #path from its neighbour to destination
            
    #Since we need all paths so mrk visited as False for backtracking and check other paths as well
    visited.remove(src)   #Postorder mein False krdo ,return time pe Visited False bana dega     
if __name__ =='__main__':
    vertices = int(input())
    graph = [ [] for _ in range(vertices) ]
    
    edges = int(input())
    for i in range(edges):
        parts = input().split(" ")
        v1 = int(parts[0])
        v2 = int(parts[1])
        wt = int(parts[2])
    
        #Since it is biderectional so '0 1 10' and '1 0 10' both we need to store
        graph[v1].append(Edge(v1,v2,wt))
        graph[v2].append(Edge(v2,v1,wt))
    
    src = int(input())
    
    visited = set()
    hamiltonian(graph,src,visited,str(src)+"",src) #send by appending src from first, so its neighbours will attach next

```

## Breadth First Search

```python
class Edge():
    def __init__(self,src,nbr,wt):
        self.src = src #Source Vertex
        self.nbr = nbr #Neighbour Vertex
        self.wt = wt #Edge Weight 
        
        
class Pair():
    def __init__(self,src,path):
        self.src = src
        self.path = path
        
if __name__ =='__main__':
    vertices = int(input())
    graph = [ [] for _ in range(vertices) ]
    
    edges = int(input())
    for i in range(edges):
        parts = input().split(" ")
        v1 = int(parts[0])
        v2 = int(parts[1])
        wt = int(parts[2])
    
        #Since it is biderectional so '0 1 10' and '1 0 10' both we need to store
        graph[v1].append(Edge(v1,v2,wt))
        graph[v2].append(Edge(v2,v1,wt))
    
    src = int(input())
    
    #BFS
    visited = [False]*vertices
    queue = []
    queue.append(Pair(src,str(src)+""))
    while(len(queue)>0):
        #remove,mark*,work,add*
        
        rem = queue.pop(0)
        #If already marked then continue,dont do any work
        if visited[rem.src] == True:
            continue
        else:
            #mark*/visit
            visited[rem.src] = True
            #work or print
            print(str(rem.src)+'@'+rem.path)
            #add*/add neighbours to the path
            for edge in graph[rem.src]:
                if visited[edge.nbr] == False:
                    queue.append(Pair(edge.nbr,rem.path+str(edge.nbr)))
```

## Shortest path in terms of Edges - BFS

```python
class Edge():
    def __init__(self,src,nbr,wt):
        self.src = src #Source Vertex
        self.nbr = nbr #Neighbour Vertex
        self.wt = wt #Edge Weight 
        
        
class Pair():
    def __init__(self,src,path):
        self.src = src
        self.path = path
        
if __name__ =='__main__':
    vertices = int(input())
    graph = [ [] for _ in range(vertices) ]
    
    edges = int(input())
    for i in range(edges):
        parts = input().split(" ")
        v1 = int(parts[0])
        v2 = int(parts[1])
        wt = int(parts[2])
    
        #Since it is biderectional so '0 1 10' and '1 0 10' both we need to store
        graph[v1].append(Edge(v1,v2,wt))
        graph[v2].append(Edge(v2,v1,wt))
    
    src = int(input())
    dst = int(input())
    
    #BFS
    res_path = []
    visited = [False]*vertices
    queue = []
    queue.append(Pair(src,str(src)+""))
    while(len(queue)>0):
        #remove,mark*,work,add*
        
        rem = queue.pop(0)
        #If already marked then continue,dont do any work
        if visited[rem.src] == True:
            continue
        else:
            #mark*/visit
            visited[rem.src] = True
            #work or print
            if rem.src == dst:
                #print(str(rem.src)+'@'+rem.path)
                res_path.append(rem.path)
                break
            #add*/add neighbours to the path
            for edge in graph[rem.src]:
                if visited[edge.nbr] == False:
                    queue.append(Pair(edge.nbr,rem.path+str(edge.nbr)))

    print(res_path)
```

## Is a Graph Cycle?

```python
class Edge():
    def __init__(self,src,nbr,wt):
        self.src = src #Source Vertex
        self.nbr = nbr #Neighbour Vertex
        self.wt = wt #Edge Weight 
class Pair():
    def __init__(self,src,path,level):
        self.src = src
        self.path = path
        self.level = level
def isBipartite(graph,src,visited): #BFS code
    
    queue = []
    queue.append(Pair(src,str(src)+"",0))
    while(len(queue)>0):
        #remove,mark*,work,add*
        
        rem = queue.pop(0)
        
        '''Cycle'''
        #If already marked then you have a cycle
        #waha aane ka koi aur rasta bhi thha
        if visited[rem.src] != -1: # visited[rem.src] == True:
            #Main logic
            #current level not equal to old level then it is odd cycle
            if rem.level != visited[rem.level]:
                return False
            else:
                return True
        else:
            #mark*/visit
            visited[rem.src] = rem.level # -- this will mark the old levels
            #work or print - No need to print here
            #print(str(rem.src)+'@'+rem.path)
            #add*/add neighbours to the path
            for edge in graph[rem.src]:
                if visited[edge.nbr] == -1:
                    queue.append(Pair(edge.nbr,rem.path+str(edge.nbr),rem.level+1))

    return False
if __name__ =='__main__':
    vertices = int(input())
    graph = [ [] for _ in range(vertices) ]
    
    edges = int(input())
    for i in range(edges):
        parts = input().split(" ")
        v1 = int(parts[0])
        v2 = int(parts[1])
        wt = int(parts[2])
    
        #Since it is biderectional so '0 1 10' and '1 0 10' both we need to store
        graph[v1].append(Edge(v1,v2,wt))
        graph[v2].append(Edge(v2,v1,wt))
    
    components = []
    result = True 
    '''We will put levels in visited, in which level it was visited'''
    visited = [-1]*vertices #using int array, -1 : not visited yet
    for v in range(vertices):
        #sirf unvisited vertex ko bhejenge cyclic check hone ke liye
        if visited[v] == -1:
            isComponentBipartite  = isBipartite(graph,v,visited)
            if isComponentBipartite == False:
                result = False
                break
    print(result)
```

## Spread Infection

```python
class Edge():
    def __init__(self,src,nbr,wt):
        self.src = src #Source Vertex
        self.nbr = nbr #Neighbour Vertex
        self.wt = wt #Edge Weight 
class Pair():
    def __init__(self,src,time):
        self.src = src
        self.time = time
def spreadInfection(graph,src,t,visited): #BFS code
    
    queue = []
    queue.append(Pair(src,1))
    count = 0
    while(len(queue)>0):
        #remove,mark*,work,add*
        
        rem = queue.pop(0)
        
        #If already marked then you have a cycle
        #waha aane ka koi aur rasta bhi thha
        if visited[rem.src] > 0:
            continue
        else:
            '''Infect people by visiting them and increase count. break if time is elapsed'''
            #mark*/visit
            visited[rem.src] = rem.time #putting non zero value/time in visited
            if rem.time>t:
                break
            #work or print 
            count+=1
            #add*/add neighbours to the path
            for edge in graph[rem.src]:
                if visited[edge.nbr] == 0: #non sick neighbours
                    #put them in queue since we have marked its source as infected
                    queue.append(Pair(edge.nbr,rem.time+1))

    return count
if __name__ =='__main__':
    vertices = int(input())
    graph = [ [] for _ in range(vertices) ]
    
    edges = int(input())
    for i in range(edges):
        parts = input().split(" ")
        v1 = int(parts[0])
        v2 = int(parts[1])
        wt = int(parts[2])
    
        #Since it is biderectional so '0 1 10' and '1 0 10' both we need to store
        graph[v1].append(Edge(v1,v2,wt))
        graph[v2].append(Edge(v2,v1,wt))
    src = int(input())
    t = int(input())
    #visited = 0 means not infected
    #visited>0 means at what time he was infected
    visited = [0]*vertices 
    
    count = spreadInfection(graph,src,t,visited)
    print(count)
```

## Dijktra's Algorithm - Single Source Shortest Path

```python
class Edge():
    def __init__(self,src,nbr,wt):
        self.src = src #Source Vertex
        self.nbr = nbr #Neighbour Vertex
        self.wt = wt #Edge Weight 
        
class Pair():
    def __init__(self,src,psf,wsf):
        self.src = src
        self.psf = psf
        self.wsf = wsf
        
    def __lt__(self,pair):
        return self.wsf < pair.wsf#priority to less weight,first element will be the smallest one
        
    
if __name__ =='__main__':
    vertices = int(input())
    graph = [ [] for _ in range(vertices) ]
    
    edges = int(input())
    for i in range(edges):
        parts = input().split(" ")
        v1 = int(parts[0])
        v2 = int(parts[1])
        wt = int(parts[2])
    
        #Since it is biderectional so '0 1 10' and '1 0 10' both we need to store
        graph[v1].append(Edge(v1,v2,wt))
        graph[v2].append(Edge(v2,v1,wt))
    
    src = int(input())
    visited = [False]*vertices
    
    import queue as Q
    pq = Q.PriorityQueue()
    pq.put(Pair(src,str(src)+"",0))
    
    while pq.qsize()>0:
        #Remove
        rem = pq.get()
        
        if visited[rem.src] == True:
            continue
           
        visited[rem.src] = True
        #Work
        print(str(rem.src)+' via '+str(rem.psf)+" @ "+str(rem.wsf))
        #add*/add neighbours to the path
        for edge in graph[rem.src]:
            if visited[edge.nbr] == False: #non sick neighbours
                #put them in queue since we have marked its source as infected
                pq.put(Pair(edge.nbr,str(rem.psf)+str(edge.nbr),rem.wsf+edge.wt))
```

## Prim's Algorithm

```python
class Edge():
    def __init__(self,src,nbr,wt):
        self.src = src #Source Vertex
        self.nbr = nbr #Neighbour Vertex
        self.wt = wt #Edge Weight 
        
class Pair():
    def __init__(self,src,av,wsf):
        self.src = src
        self.av = av
        self.wsf = wsf
        
    def __lt__(self,pair):
        return self.wsf < pair.wsf#priority to less weight,first element will be the smallest one
        
    
if __name__ =='__main__':
    vertices = int(input())
    graph = [ [] for _ in range(vertices) ]
    
    edges = int(input())
    for i in range(edges):
        parts = input().split(" ")
        v1 = int(parts[0])
        v2 = int(parts[1])
        wt = int(parts[2])
    
        #Since it is biderectional so '0 1 10' and '1 0 10' both we need to store
        graph[v1].append(Edge(v1,v2,wt))
        graph[v2].append(Edge(v2,v1,wt))
    
    src = int(input())
    visited = [False]*vertices
    
    import queue as Q
    pq = Q.PriorityQueue()
    pq.put(Pair(src,-1,0))
    
    while pq.qsize()>0:
        #Remove
        rem = pq.get()
        
        if visited[rem.src] == True:
            continue
           
        visited[rem.src] = True
        #Work
        if rem.av!=-1:
            print(str(rem.src)+' - '+str(rem.av)+" @ "+str(rem.wsf))
        #add*/add neighbours to the path
        for edge in graph[rem.src]:
            if visited[edge.nbr] == False: #non sick neighbours
                pq.put(Pair(edge.nbr,rem.src,edge.wt))
        
```

## Topological Sort - Order of work

```python
class Edge():
    def __init__(self,src,nbr):
        self.src = src #Source Vertex
        self.nbr = nbr #Neighbour Vertex
        
def topologicalSort(graph,src,visited,stack):
    visited[src] = True
    #Now keep asking the neighours
    for edge in graph[src]: #Accessing list inside a list
        if visited[edge.nbr] == False: #if neighbor is not visited yet then call its neighbor
            topologicalSort(graph,edge.nbr,visited,stack) 
    #Postorder        
    stack.append(src)
        
    
if __name__ =='__main__':
    vertices = int(input())
    graph = [ [] for _ in range(vertices) ]
    
    edges = int(input())
    for i in range(edges):
        parts = input().split(" ")
        v1 = int(parts[0])
        v2 = int(parts[1])
    
        #Since it is uni-directional so '0 1 10' and '1 0 10' both we need to store
        graph[v1].append(Edge(v1,v2))
    
    visited = [False]*vertices
    stack = []
    
    for v in range(vertices):
        if visited[v] == False:
            topologicalSort(graph,v,visited,stack)
    print('Topological Sort : ',stack[::-1])
    print('Order of work : ',stack)
```
