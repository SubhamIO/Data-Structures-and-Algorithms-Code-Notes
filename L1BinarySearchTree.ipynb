## Display a Binary Search Tree

```python
class Node():
 
    def __init__(self, data,left,right):
        self.data = data
        self.left = left
        self.right = right

def construct(arr,lo,hi):
    if lo>hi:
        return None
    mid = (lo+hi)//2
    data = arr[mid]
    left_child = construct(arr,lo,mid-1)
    right_child = construct(arr,mid+1,hi)
    
    node = Node(data,left_child,right_child)
    
    return node
    

'''
- EXP : display(10) will print itself and its family
- FAITH : display(20),display(30),display(40) will print themselves and their families
- EXP-FAITH : display(10) = self(10)+display(20)+display(30)+display(40)'''
#If I send Node(10) as input, it will print
def display(node):
    #BASE CASE
    #When there is no children node then return
    if node == None:
        return 
    
    string = ""    
    string+= "." if node.left==None else str(node.left.data) + ""
    string+= "<--"+str(node.data) + "-->"
    string+= "." if node.right == None else str(node.right.data) + ""    
    print(string)
    
    #FAITH
    display(node.left)
    display(node.right)

if __name__ == '__main__':
    arr = [12,25,37,50,62,75,87]
    low = 0
    high = len(arr)-1
    root = construct(arr, low, high)

    display(root)
  
```

## Size, Sum, Max,Min,Height,Find

```python
class Node():
 
    def __init__(self, data,left,right):
        self.data = data
        self.left = left
        self.right = right

def construct(arr,lo,hi):
    if lo>hi:
        return None
    mid = (lo+hi)//2
    data = arr[mid]
    left_child = construct(arr,lo,mid-1)
    right_child = construct(arr,mid+1,hi)
    
    node = Node(data,left_child,right_child)
    
    return node
    
def size(node):
    if node==None:
        return 0
    left_size = size(node.left)
    right_size = size(node.right)
    
    return left_size+right_size+1

def sumOfNodes(node):
    #Base Case
    #If no nodes then sum = 0 
    if node == None:
        return 0
    
    left_child_sum = sumOfNodes(node.left)
    right_child_sum = sumOfNodes(node.right)
    total_sum = left_child_sum+right_child_sum + node.data
    
    return total_sum

def max_finder(node): 
    if node.right!=None:
        return max_finder(node.right)
    else:
        return node.data #last element to the right will be largest in BST

def min_finder(node): 
    if node.left!=None:
        return min_finder(node.left)
    else:
        return node.data #last element to the left will be smallest in BST

def height_finder(node):   
    #Base Case
    #ht = -1 : Edge wise depth
    #ht = 0 : Node wise depth(1 more than Edge wise depth)
    if node == None:
        return -1
        
    left_child_ht = height_finder(node.left)
    right_child_ht = height_finder(node.right)
    ht = max(left_child_ht,right_child_ht) + 1  
        
    return ht


def find(node,data):
    if node ==None:
        return False
    
    if data>node.data:
        return find(node.right,data)
    
    elif data<node.data:
        return find(node.left,data)
    
    else:
        return True #node.data==data then return True
if __name__ == '__main__':
    arr = [12,25,37,50,62,75,87]
    low = 0
    high = len(arr)-1
    root = construct(arr, low, high)

    sz = size(root)
    sumOfnodes = sumOfNodes(root)
    max_value = max_finder(root)
    min_value = min_finder(root)
    heightOfBT = height_finder(root)
    f = find(root,87)
    print('Size: ',sz)
    print('Sum: ',sumOfnodes)
    print('MAX value: ',max_value)
    print('MIN value: ',min_value)
    print('Height of Binary Tree :',heightOfBT)
    print('Found: ',f)
```

## Add a node in BST

```python
class Node():
 
    def __init__(self, data,left,right):
        self.data = data
        self.left = left
        self.right = right

def construct(arr,lo,hi):
    if lo>hi:
        return None
    mid = (lo+hi)//2
    data = arr[mid]
    left_child = construct(arr,lo,mid-1)
    right_child = construct(arr,mid+1,hi)
    
    node = Node(data,left_child,right_child)
    
    return node

def display(node):
    #BASE CASE
    #When there is no children node then return
    if node == None:
        return 
    
    string = ""    
    string+= "." if node.left==None else str(node.left.data) + ""
    string+= "<--"+str(node.data) + "-->"
    string+= "." if node.right == None else str(node.right.data) + ""    
    print(string)
    
    #FAITH
    display(node.left)
    display(node.right)
    
def addNode(node,data):
    #node --None pohch gaya toh Node banao(data) ka and attach it to right/left depending on data
    if node==None:
        return Node(data,None,None)
    
    if data>node.data:
        node.right = addNode(node.right,data) #attaching the Node(data) to right if data was larger
    elif data<node.data:  
        node.left = addNode(node.left,data) #attaching the Node(data) to left if data was smaller
    return node
        
    

if __name__ == '__main__':
    arr = [12,25,37,50,62,75,87]
    low = 0
    high = len(arr)-1
    root = construct(arr, low, high)
    addNode(root,90)
    display(root)
    

    
```

## Remove a Node in BST

```python
class Node():
 
    def __init__(self, data,left,right):
        self.data = data
        self.left = left
        self.right = right

def construct(arr,lo,hi):
    if lo>hi:
        return None
    mid = (lo+hi)//2
    data = arr[mid]
    left_child = construct(arr,lo,mid-1)
    right_child = construct(arr,mid+1,hi)
    
    node = Node(data,left_child,right_child)
    
    return node

def display(node):
    #BASE CASE
    #When there is no children node then return
    if node == None:
        return 
    
    string = ""    
    string+= "." if node.left==None else str(node.left.data) + ""
    string+= "<--"+str(node.data) + "-->"
    string+= "." if node.right == None else str(node.right.data) + ""    
    print(string)
    
    #FAITH
    display(node.left)
    display(node.right)
    

def max_finder(node): 
    if node.right!=None:
        return max_finder(node.right)
    else:
        return node.data #last element to the right will be largest in BST        
def removeNode(node,data):
    if node==None:
        return None
    if data>node.data:
        node.right = removeNode(node.right,data)
    elif data<node.data:
        node.left = removeNode(node.left,data)
    else: #when data == node.data(found the node to delete)
        #Have 2 child
        if node.left!=None and node.right!=None:
            max_val = max_finder(node.left) #Find max from left subtree
            node.data = max_val #replace nodes current value with max_val from left subtree
            node.left = removeNode(node.left,max_val) #remove the Node(max_val) from left subtree
            return node
        #Have only left child
        elif node.left!=None:
            return node.left
        #Have only right child
        elif node.right!=None:
            return node.right
        #Have no child
        else:
            return None
    return node    

if __name__ == '__main__':
    arr = [12,25,37,50,62,75,87]
    low = 0
    high = len(arr)-1
    root = construct(arr, low, high)
    res = removeNode(root,75)
    display(res)
    

    
```

## Replace Sum of Larger in BST

```python
class Node():
 
    def __init__(self, data,left,right):
        self.data = data
        self.left = left
        self.right = right

def construct(arr,lo,hi):
    if lo>hi:
        return None
    mid = (lo+hi)//2
    data = arr[mid]
    left_child = construct(arr,lo,mid-1)
    right_child = construct(arr,mid+1,hi)
    
    node = Node(data,left_child,right_child)
    
    return node

def display(node):
    #BASE CASE
    #When there is no children node then return
    if node == None:
        return 
    
    string = ""    
    string+= "." if node.left==None else str(node.left.data) + ""
    string+= "<--"+str(node.data) + "-->"
    string+= "." if node.right == None else str(node.right.data) + ""    
    print(string)
    
    #FAITH
    display(node.left)
    display(node.right)
    

total_sum = 0
def replaceSumOfLarger(node):
    global total_sum
    if node == None:
        return
    #Reverse Inorder, so traverse right side first
    replaceSumOfLarger(node.right)
    
    #Inorder area
    original_data = node.data
    node.data = total_sum
    total_sum+=original_data
    
    replaceSumOfLarger(node.left)

if __name__ == '__main__':
    arr = [12,25,37,50,62,75,87]
    low = 0
    high = len(arr)-1
    root = construct(arr, low, high)
    replaceSumOfLarger(root)
    display(root)
    
```

## LCA of BST

```python
class Node():
 
    def __init__(self, data,left,right):
        self.data = data
        self.left = left
        self.right = right

def construct(arr,lo,hi):
    if lo>hi:
        return None
    mid = (lo+hi)//2
    data = arr[mid]
    left_child = construct(arr,lo,mid-1)
    right_child = construct(arr,mid+1,hi)
    
    node = Node(data,left_child,right_child)
    
    return node

def display(node):
    #BASE CASE
    #When there is no children node then return
    if node == None:
        return 
    
    string = ""    
    string+= "." if node.left==None else str(node.left.data) + ""
    string+= "<--"+str(node.data) + "-->"
    string+= "." if node.right == None else str(node.right.data) + ""    
    print(string)
    
    #FAITH
    display(node.left)
    display(node.right)
    

def lcaBST(node,d1,d2):
    if node == None:
        return
    if d1<node.data and d2<node.data:
        return lcaBST(node.left,d1,d2)
    elif d1>node.data and d2>node.data:
        return lcaBST(node.right,d1,d2) 
    else:
        return node.data

if __name__ == '__main__':
    arr = [12,25,30,37,40,50,60,62,70,75,87]
    low = 0
    high = len(arr)-1
    root = construct(arr, low, high)
    res = lcaBST(root,62,87)
    print('LCA : ',res)
    
```

## Target Sum Pair

```python
class Node():
 
    def __init__(self, data,left,right):
        self.data = data
        self.left = left
        self.right = right

def construct(arr,lo,hi):
    if lo>hi:
        return None
    mid = (lo+hi)//2
    data = arr[mid]
    left_child = construct(arr,lo,mid-1)
    right_child = construct(arr,mid+1,hi)
    
    node = Node(data,left_child,right_child)
    
    return node

def display(node):
    #BASE CASE
    #When there is no children node then return
    if node == None:
        return 
    
    string = ""    
    string+= "." if node.left==None else str(node.left.data) + ""
    string+= "<--"+str(node.data) + "-->"
    string+= "." if node.right == None else str(node.right.data) + ""    
    print(string)
    
    #FAITH
    display(node.left)
    display(node.right)
    
def find(node,data):
    if node ==None:
        return False
    
    if data>node.data:
        return find(node.right,data)
    
    elif data<node.data:
        return find(node.left,data)
    
    else:
        return True #node.data==data then return True
    
def targetSumPair(root,node,target):
    if node == None:
        return
    targetSumPair(root,node.left,target)
    
    complement = target - node.data
    if node.data<complement: 
        #To find an element in BST we need root node, so passed root as a parameter
        if find(root,complement) == True:
            print(str(node.data)+'-'+str(complement))
    
    
    targetSumPair(root,node.right,target)
        

if __name__ == '__main__':
    arr = [12,25,30,37,40,50,60,62,70,75,87]
    low = 0
    high = len(arr)-1
    root = construct(arr, low, high)
    target = 100
    targetSumPair(root,root,target)
    
```

## Leetcode Solution - Target Sum Pair - True/False

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def findTarget(self, root: TreeNode, k: int) -> bool:
        def find(node,data):
            if node ==None:
                return False

            if data>node.val:
                return find(node.right,data)

            elif data<node.val:
                return find(node.left,data)

            else:
                return True

        def targetSumPair(root,node,target):
            if node == None:
                return
            targetSumPair(root,node.left,target)

            complement = target - node.val
            if node.val<complement: 
                
                if find(root,complement) == True:
                    self.res = True


            targetSumPair(root,node.right,target)
        self.res = False
        targetSumPair(root,root,k)
        return self.res
```

## Optimised Approach - Target Sum Pair
- Time = O(n) , Space= O(logn)

```python
class Node():
 
    def __init__(self, data,left,right):
        self.data = data
        self.left = left
        self.right = right

def construct(arr,lo,hi):
    if lo>hi:
        return None
    mid = (lo+hi)//2
    data = arr[mid]
    left_child = construct(arr,lo,mid-1)
    right_child = construct(arr,mid+1,hi)
    
    node = Node(data,left_child,right_child)
    
    return node

def display(node):
    #BASE CASE
    #When there is no children node then return
    if node == None:
        return 
    
    string = ""    
    string+= "." if node.left==None else str(node.left.data) + ""
    string+= "<--"+str(node.data) + "-->"
    string+= "." if node.right == None else str(node.right.data) + ""    
    print(string)
    
    #FAITH
    display(node.left)
    display(node.right)
    
def find(node,data):
    if node ==None:
        return False
    
    if data>node.data:
        return find(node.right,data)
    
    elif data<node.data:
        return find(node.left,data)
    
    else:
        return True #node.data==data then return True
    
class IterativePair(): 
    def __init__(self, node,state):
        self.node = node
        self.state = state 
        
def getNextFromNormalInorder(stack):
    while len(stack) > 0:
        #pick the top element Pair
        top = stack[-1]
        
        if top.state == 1: #Pre , State++ , Left
            top.state+=1
            if top.node.left != None:
                stack.append(IterativePair(top.node.left,1))
            
        elif top.state == 2:#In , State++ , Right
            top.state+=1
            if top.node.right != None:
                stack.append(IterativePair(top.node.right,1))
            return top.node
        
        else: #Post, pop
            stack.pop()          
    return None 
        
def getNextFromReverseInorder(stack):
    while len(stack) > 0:
        #pick the top element Pair
        top = stack[-1]
        
        if top.state == 1: #Pre , State++ , Right
            top.state+=1
            if top.node.right != None:
                stack.append(IterativePair(top.node.right,1))
            
        elif top.state == 2:#In , State++ , Left
            top.state+=1
            if top.node.left != None:
                stack.append(IterativePair(top.node.left,1))
            return top.node
        
        else: #Post, pop
            stack.pop()          
    return None
    
def iterativePreorderAndPostorder(node,target):
    

    left_stack = []
    right_stack = []
    left_stack.append(IterativePair(node,1)) 
    right_stack.append(IterativePair(node,1)) 
    
    #Get single elements from both normal and reverse inorder
    ascending_inorder = getNextFromNormalInorder(left_stack)
    descending_inorder = getNextFromReverseInorder(right_stack)
    
    while ascending_inorder.data<descending_inorder.data: #n
        #get next element from normal inorder when it is less than target
        #like increasing low+=1 pointer
        if ascending_inorder.data+descending_inorder.data<target:
            ascending_inorder = getNextFromNormalInorder(left_stack)
            
        #get next element from reverse inorder when it is more than target
        #like decreasing high-=1 pointer   
        elif ascending_inorder.data+descending_inorder.data>target:
            descending_inorder = getNextFromReverseInorder(right_stack)
        
        #get next element from normal and reverse inorder when it is equal to the target
        #like increasing low+=1 and decreasing high-=1 pointer
        else:
            print(ascending_inorder.data,descending_inorder.data)
            ascending_inorder = getNextFromNormalInorder(left_stack)
            descending_inorder = getNextFromReverseInorder(right_stack)
    
    

if __name__ == '__main__':
    arr = [12,25,30,37,40,50,60,62,70,75,87]
    low = 0
    high = len(arr)-1
    root = construct(arr, low, high)
    target = 100
    #inorder_list = iterativePreorderAndPostorder(root)
    iterativePreorderAndPostorder(root,target)
    
    
```
