```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
   
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
            
        carry = 0
        index = 0
        result = ListNode()
        while l1 is not None or l2 is not None or carry >  0:
            sumval = carry
            if l1 is not None:
                sumval += l1.val
                l1 = l1.next
            if l2 is not None:
                sumval += l2.val
                l2 = l2.next
            carry = sumval // 10
            sumval = sumval % 10
            if index == 0:
                n = result
                n.val = sumval
                index += 1
            else:
                n.next = ListNode(sumval)
                n = n.next
        return result
```

## Interection of LL - Leetcode Soln

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> ListNode:
        def sizer(head):
            ct= 0
            node = head
            while node!=None:
                node=node.next
                ct+=1
            return ct
        
        def findIntersection(l1,l2):
            t1 = l1
            t2 = l2

            gap = abs(sizer(l1)-sizer(l2))
            if sizer(l1)>sizer(l2):
                for i in range(gap):
                    t1 = t1.next
            else:
                for i in range(gap):
                    t2= t2.next
            
            while t1!=t2:
                t1 = t1.next
                t2 = t2.next
            
            return t1
        inter = findIntersection(headA,headB)
        return inter
```

## RemoveElements :   Leetcode accepted solution 
- Input: head = [1,2,6,3,4,5,6], val = 6
- Output: [1,2,3,4,5]

```python
class Solution:
    def removeElements(self, head: ListNode, val: int) -> ListNode:
        if head==None: 
            return
        if head.val == val: 
             return self.removeElements(head.next, val)
        else: 
            head.next = self.removeElements(head.next, val)
        return head
```
