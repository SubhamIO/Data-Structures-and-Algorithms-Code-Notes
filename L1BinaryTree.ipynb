## Display Binary Tree

```python
class Node():
 
    def __init__(self, data,left,right):
        self.data = data
        self.left = left
        self.right = right
class Pair(): 
    def __init__(self, node,state):
        self.node = node
        self.state = state 

def construct(arr):
    stack = []
    root = Node(arr[0],None,None)
    stack.append(Pair(root,1))
    
    i = 0
    while len(stack) > 0:       
        #pick the top element Pair
        top = stack[-1]
        if top.state == 1:
            i+=1
            if arr[i] != None:
                left_node = Node(arr[i],None,None)
                top.node.left = left_node
                stack.append(Pair(left_node,1))
                top.state+=1
            else:
                top.node.left = None
                top.state+=1
                
        elif top.state == 2:
            i+=1
            if arr[i] != None:
                right_node = Node(arr[i],None,None)
                top.node.right = right_node
                stack.append(Pair(right_node,1))
                top.state+=1
            else:
                top.node.right = None
                top.state+=1
        else:
            stack.pop()
    return root


'''
- EXP : display(10) will print itself and its family
- FAITH : display(20),display(30),display(40) will print themselves and their families
- EXP-FAITH : display(10) = self(10)+display(20)+display(30)+display(40)'''
#If I send Node(10) as input, it will print
def display(node):
    #BASE CASE
    #When there is no children node then return
    if node == None:
        return 
    
    string = ""    
    string+= "." if node.left==None else str(node.left.data) + ""
    string+= "<--"+str(node.data) + "-->"
    string+= "." if node.right == None else str(node.right.data) + ""    
    print(string)
    
    #FAITH
    display(node.left)
    display(node.right)

if __name__ == '__main__':
    arr = [50,25,12,None,None,37,30,None,None,None,75,62,None,70,None,None,87,None,None]
    root = construct(arr)

    display(root)
  
```

## Size,Sum,Max Element,Height of a Binary Tree
- Remember : In base case return what you should have initialised first, 
- like s=0 for sum, max = float('-inf') for max ,height = -1 .. etc

```python
class Node():
 
    def __init__(self, data,left,right):
        self.data = data
        self.left = left
        self.right = right
class Pair(): 
    def __init__(self, node,state):
        self.node = node
        self.state = state 

def construct(arr):
    stack = []
    root = Node(arr[0],None,None)
    stack.append(Pair(root,1))
    
    i = 0
    while len(stack) > 0:       
        #pick the top element Pair
        top = stack[-1]
        if top.state == 1:
            i+=1
            if arr[i] != None:
                left_node = Node(arr[i],None,None)
                top.node.left = left_node
                stack.append(Pair(left_node,1))
                top.state+=1
            else:
                top.node.left = None
                top.state+=1
                
        elif top.state == 2:
            i+=1
            if arr[i] != None:
                right_node = Node(arr[i],None,None)
                top.node.right = right_node
                stack.append(Pair(right_node,1))
                top.state+=1
            else:
                top.node.right = None
                top.state+=1
        else:
            stack.pop()
    return root


'''
- EXP : display(10) will print itself and its family
- FAITH : display(20),display(30),display(40) will print themselves and their families
- EXP-FAITH : display(10) = self(10)+display(20)+display(30)+display(40)'''
#If I send Node(10) as input, it will print
def display(node):
    #BASE CASE
    #When there is no children node then return
    if node == None:
        return 
    
    string = ""    
    string+= "." if node.left==None else str(node.left.data) + ""
    string+= "<--"+str(node.data) + "-->"
    string+= "." if node.right == None else str(node.right.data) + ""    
    print(string)
    
    #FAITH
    display(node.left)
    display(node.right)
    
def size(node):
    #Base Case
    #If no nodes then size = 0(dosnt increase)
    if node == None:
        return 0
    
    left_child_size = size(node.left)
    right_child_size = size(node.right)
    total_size = left_child_size+right_child_size + 1
    
    return total_size

def sumOfNodes(node):
    #Base Case
    #If no nodes then sum = 0 
    if node == None:
        return 0
    
    left_child_sum = sumOfNodes(node.left)
    right_child_sum = sumOfNodes(node.right)
    total_sum = left_child_sum+right_child_sum + node.data
    
    return total_sum

def max_finder(node): 
    #Base Case
    if node == None:
        return float('-inf')
    
    #Here we will get the max node from the childs
    max_val = float('-inf')
    left_child_max = max_finder(node.left)
    right_child_max = max_finder(node.right)
    
    max_val = max(node.data,max(left_child_max,right_child_max))
    
    return max_val

def height_finder(node):   
    #Base Case
    #ht = -1 : Edge wise depth
    #ht = 0 : Node wise depth(1 more than Edge wise depth)
    if node == None:
        return -1
    
    
    
    left_child_ht = height_finder(node.left)
    right_child_ht = height_finder(node.right)
    ht = max(left_child_ht,right_child_ht) + 1  
        
    return ht

if __name__ == '__main__':
    arr = [50,25,12,None,None,37,30,None,None,None,75,62,None,70,None,None,87,None,None]
    root = construct(arr)

    sz = size(root)
    sumOfnodes = sumOfNodes(root)
    max_value = max_finder(root)
    heightOfBT = height_finder(root)
    print('Size :',sz)
    print('Sum :',sumOfnodes)
    print('Max Element',max_value)
    print('Height of Binary Tree :',heightOfBT)
```

```python
class Node():
 
    def __init__(self, data,left,right):
        self.data = data
        self.left = left
        self.right = right
class Pair(): 
    def __init__(self, node,state):
        self.node = node
        self.state = state 

def construct(arr):
    stack = []
    root = Node(arr[0],None,None)
    stack.append(Pair(root,1))
    
    i = 0
    while len(stack) > 0:       
        #pick the top element Pair
        top = stack[-1]
        if top.state == 1:
            i+=1
            if arr[i] != None:
                left_node = Node(arr[i],None,None)
                top.node.left = left_node
                stack.append(Pair(left_node,1))
                top.state+=1
            else:
                top.node.left = None
                top.state+=1
                
        elif top.state == 2:
            i+=1
            if arr[i] != None:
                right_node = Node(arr[i],None,None)
                top.node.right = right_node
                stack.append(Pair(right_node,1))
                top.state+=1
            else:
                top.node.right = None
                top.state+=1
        else:
            stack.pop()
    return root


def levelOrderLinewiseZigZag(node):
    if node == None:
        return 
    
    #using 2 queues
    ms = [] #main queue
    cs = [] #child queue
    #First add the Root node to Main Queue
    ms.append(node)
    level = 1 #make this 0 to get opposite zip
    while(len(ms) > 0):
        #Printing top to bottom of stack
        node = ms.pop() #pop last element
        print(str(node.data),end = " ")
        
        #The direction in which you enter children to child stack,
        #They will be later printed in opposite direction
        if level%2 != 0:
            #for i in range (len(node.children)):
            if node.left!=None:
                cs.append(node.left)
            if node.right!=None:
                cs.append(node.right)
        else:
            #for i in range(len(node.children)-1,-1,-1):
            
            if node.right!=None:
                cs.append(node.right)
            if node.left!=None:
                cs.append(node.left)
            
        #New Level   
        if len(ms) == 0:
            level +=1
            ms = cs
            cs = []
            print()
            
def levelOrderLinewiseCountApproach(node):
    if node == None:
        return ""
    #using 1 queue1
    mq = [] #main queue
    #First add the Root node to Main Queue
    mq.append(node)
    
    while(len(mq) > 0):
        ct = len(mq)
        for i in range(ct):    
            node = mq.pop(0) #Remove
            print(node.data,end = " ") #Print
            if node.left != None:
                mq.append(node.left) #Add the child
            if node.right != None:
                mq.append(node.right)
            
        print()  

if __name__ == '__main__':
    arr = [50,25,12,None,None,37,30,None,None,None,75,62,None,70,None,None,87,None,None]
    root = construct(arr)
    
    levelOrderLinewiseCountApproach(root)
    print('*'*10)
    levelOrderLinewiseZigZag(root)
```

## Iterative Preorder,Inorder and Postorder

```python
class Node():
 
    def __init__(self, data,left,right):
        self.data = data
        self.left = left
        self.right = right
class Pair(): 
    def __init__(self, node,state):
        self.node = node
        self.state = state 

def construct(arr):
    stack = []
    root = Node(arr[0],None,None)
    stack.append(Pair(root,1))
    
    i = 0
    while len(stack) > 0:       
        #pick the top element Pair
        top = stack[-1]
        if top.state == 1:
            i+=1
            if arr[i] != None:
                left_node = Node(arr[i],None,None)
                top.node.left = left_node
                stack.append(Pair(left_node,1))
                top.state+=1
            else:
                top.node.left = None
                top.state+=1
                
        elif top.state == 2:
            i+=1
            if arr[i] != None:
                right_node = Node(arr[i],None,None)
                top.node.right = right_node
                stack.append(Pair(right_node,1))
                top.state+=1
            else:
                top.node.right = None
                top.state+=1
        else:
            stack.pop()
    return root


    
class Pair(): 
    def __init__(self, node,state):
        self.node = node
        self.state = state       
def iterativePreorderAndPostorder(node):
    
    if node == None:
        return
    stack = []
    stack.append(Pair(node,1)) #push node to stack, with state = -1
    
    pre = ""
    post = ""
    inordr = ""
    
    while len(stack) > 0:
        #pick the top element Pair
        top = stack[-1]
        
        if top.state == 1: #Pre , State++ , Left
            pre+=str(top.node.data) + " "
            top.state+=1
            if top.node.left != None:
                stack.append(Pair(top.node.left,1))
            
        elif top.state == 2:#In , State++ , Right
            inordr+=str(top.node.data) + " "
            top.state+=1
            if top.node.right != None:
                stack.append(Pair(top.node.right,1))
       
        
        else: #Post, pop
            post+=str(top.node.data) + " "
            stack.pop()            
    print('Preorder :',pre)
    print('Inorder :',inordr)
    print('Postorder :',post)  

if __name__ == '__main__':
    arr = [50,25,12,None,None,37,30,None,None,None,75,62,None,70,None,None,87,None,None]
    root = construct(arr)
    
    iterativePreorderAndPostorder(root)
```

## Find and Node to Root Path

```python
class Node():
 
    def __init__(self, data,left,right):
        self.data = data
        self.left = left
        self.right = right
class Pair(): 
    def __init__(self, node,state):
        self.node = node
        self.state = state 

def construct(arr):
    stack = []
    root = Node(arr[0],None,None)
    stack.append(Pair(root,1))
    
    i = 0
    while len(stack) > 0:       
        #pick the top element Pair
        top = stack[-1]
        if top.state == 1:
            i+=1
            if arr[i] != None:
                left_node = Node(arr[i],None,None)
                top.node.left = left_node
                stack.append(Pair(left_node,1))
                top.state+=1
            else:
                top.node.left = None
                top.state+=1
                
        elif top.state == 2:
            i+=1
            if arr[i] != None:
                right_node = Node(arr[i],None,None)
                top.node.right = right_node
                stack.append(Pair(right_node,1))
                top.state+=1
            else:
                top.node.right = None
                top.state+=1
        else:
            stack.pop()
    return root


    
def findAndNodetoRootPath(node,data):
    if node == None:
        return False
    
    #If the root is the element or not
    if node.data == data:
        path = []
        path.append(node.data)
        return True,path
    #If element is not root, then ask children 
    if node.left!=None:
        path_till_child = []
        inleft,pathleft = findAndNodetoRootPath(node.left,data)
        path_till_child.extend(pathleft)
        if inleft == True and len(path_till_child) > 0:
            path_till_child.append(node.data)
            return True,path_till_child
        
    if node.right!=None:
        path_till_child = []
        inright,pathright = findAndNodetoRootPath(node.right,data)
        path_till_child.extend(pathright)
        if inright == True and len(path_till_child) > 0:
            path_till_child.append(node.data)
            return True,path_till_child
    
    #Not there in root, not there in child node subtree, so return False
    return False,[]
    
if __name__ == '__main__':
    arr = [50,25,12,None,None,37,30,None,None,None,75,62,None,70,None,None,87,None,None]
    root = construct(arr)
    
    res,path = findAndNodetoRootPath(root,70)
    print(res)
    print('Path : ',path)
```

## Tilt of a Binary Tree
- Travel and Chnage strategy
- Return sum, chnage tilt

```python
class Node():
 
    def __init__(self, data,left,right):
        self.data = data
        self.left = left
        self.right = right
class Pair(): 
    def __init__(self, node,state):
        self.node = node
        self.state = state 

def construct(arr):
    stack = []
    root = Node(arr[0],None,None)
    stack.append(Pair(root,1))
    
    i = 0
    while len(stack) > 0:       
        #pick the top element Pair
        top = stack[-1]
        if top.state == 1:
            i+=1
            if arr[i] != None:
                left_node = Node(arr[i],None,None)
                top.node.left = left_node
                stack.append(Pair(left_node,1))
                top.state+=1
            else:
                top.node.left = None
                top.state+=1
                
        elif top.state == 2:
            i+=1
            if arr[i] != None:
                right_node = Node(arr[i],None,None)
                top.node.right = right_node
                stack.append(Pair(right_node,1))
                top.state+=1
            else:
                top.node.right = None
                top.state+=1
        else:
            stack.pop()
    return root

#Return Sum, Change tilt

def tiltOfBt(node):
    global tilt
    if(node == None):
        return 0

    #will return left's sum and change tilt for left side
    left_sum = tiltOfBt(node.left)
    #will return right's sum and change tilt for right side
    right_sum = tiltOfBt(node.right)
    
    tilt+= abs(left_sum - right_sum)

    return left_sum+right_sum+node.data
if __name__ == '__main__':
    arr = [50,25,12,None,None,37,30,None,None,None,75,62,None,70,None,None,87,None,None]
    root = construct(arr) 
    tilt = 0
    tiltOfBt(root) 
    print('Tilt: ',tilt)
```

```python
class Node():
 
    def __init__(self, data,left,right):
        self.data = data
        self.left = left
        self.right = right
class Pair(): 
    def __init__(self, node,state):
        self.node = node
        self.state = state 

def construct(arr):
    stack = []
    root = Node(arr[0],None,None)
    stack.append(Pair(root,1))
    
    i = 0
    while len(stack) > 0:       
        #pick the top element Pair
        top = stack[-1]
        if top.state == 1:
            i+=1
            if arr[i] != None:
                left_node = Node(arr[i],None,None)
                top.node.left = left_node
                stack.append(Pair(left_node,1))
                top.state+=1
            else:
                top.node.left = None
                top.state+=1
                
        elif top.state == 2:
            i+=1
            if arr[i] != None:
                right_node = Node(arr[i],None,None)
                top.node.right = right_node
                stack.append(Pair(right_node,1))
                top.state+=1
            else:
                top.node.right = None
                top.state+=1
        else:
            stack.pop()
    return root

class BSTPair():
    def __init__(self,isBST,min_val,max_val):
        self.isBST = isBST
        self.min_val = min_val
        self.max_val = max_val
def isBST(node):
    if node == None:
        return BSTPair(True,float('inf'),float('-inf'))
    #We are getting 3 info using Pair
    left_pair = isBST(node.left)
    right_pair = isBST(node.right)
    
    #tBST of left, tBST of right, node to be BST
    own_tBST = left_pair.isBST and right_pair.isBST and (node.data>left_pair.max_val and node.data<right_pair.min_val)
    #own min and own max
    own_min = min(node.data, min(left_pair.min_val , right_pair.min_val))
    own_max = max(node.data, max(left_pair.max_val , right_pair.max_val))
    
    my_pair = BSTPair(own_tBST,own_min,own_max)
    
    return my_pair
#Inorder approach : Any Binary Tree Inorder is Sorted - Then it is BST
prev = None
def isBST2(node):
    global prev
    if node==None:
        return True
    
    if (isBST2(node.left)!=True):
        return False
    #INORDER AREA
    if prev != None and prev.data>root.data:#checking using 2 pointer approach to check sorted list or not
        return False
    prev = node
    
    
    if (isBST2(node.right)!=True):
        return False
    return True
    

if __name__ == '__main__':
    arr = [50,25,90,None,None,37,30,None,None,None,75,62,None,70,None,None,87,None,None]
    root = construct(arr) 
    own_res = isBST(root) 
    print('Is BST : ',own_res.isBST)
    
    own_res = isBST2(root) 
    print('Is BST : ',own_res)
```

```python
class Node():
 
    def __init__(self, data,left,right):
        self.data = data
        self.left = left
        self.right = right
class Pair(): 
    def __init__(self, node,state):
        self.node = node
        self.state = state 

def construct(arr):
    stack = []
    root = Node(arr[0],None,None)
    stack.append(Pair(root,1))
    
    i = 0
    while len(stack) > 0:       
        #pick the top element Pair
        top = stack[-1]
        if top.state == 1:
            i+=1
            if arr[i] != None:
                left_node = Node(arr[i],None,None)
                top.node.left = left_node
                stack.append(Pair(left_node,1))
                top.state+=1
            else:
                top.node.left = None
                top.state+=1
                
        elif top.state == 2:
            i+=1
            if arr[i] != None:
                right_node = Node(arr[i],None,None)
                top.node.right = right_node
                stack.append(Pair(right_node,1))
                top.state+=1
            else:
                top.node.right = None
                top.state+=1
        else:
            stack.pop()
    return root

#Return Sum, Change tilt

def isBalanced(node):
    global isBal
    if(node == None):
        return -1

    left_ht = isBalanced(node.left)
    right_ht = isBalanced(node.right)
    
    if abs(left_ht-right_ht)>1:
        isBal = False


    return max(left_ht,right_ht)+1
if __name__ == '__main__':
#     arr = [50,25,12,None,None,37,30,None,None,None,75,62,None,70,None,None,87,None,None]
    arr = [50,25,12,10,None,None,None,None,6,None,7,None,9,None,None]
    root = construct(arr) 
    isBal = True
    isBalanced(root) 
    print('Is Balanced : ',isBal)
```

```python
class Node():
 
    def __init__(self, data,left,right):
        self.data = data
        self.left = left
        self.right = right
class Pair(): 
    def __init__(self, node,state):
        self.node = node
        self.state = state 

def construct(arr):
    stack = []
    root = Node(arr[0],None,None)
    stack.append(Pair(root,1))
    
    i = 0
    while len(stack) > 0:       
        #pick the top element Pair
        top = stack[-1]
        if top.state == 1:
            i+=1
            if arr[i] != None:
                left_node = Node(arr[i],None,None)
                top.node.left = left_node
                stack.append(Pair(left_node,1))
                top.state+=1
            else:
                top.node.left = None
                top.state+=1
                
        elif top.state == 2:
            i+=1
            if arr[i] != None:
                right_node = Node(arr[i],None,None)
                top.node.right = right_node
                stack.append(Pair(right_node,1))
                top.state+=1
            else:
                top.node.right = None
                top.state+=1
        else:
            stack.pop()
    return root

#Return Sum, Change tilt
class BalPair():
    def __init__(self,ht,isBal):
        self.ht = ht
        self.isBal = isBal
def isBalanced(node):
    if(node == None):
        return BalPair(-1,True)

    left_pair = isBalanced(node.left)
    right_pair = isBalanced(node.right)
    
    #Own data pair
    own_ht = max(left_pair.ht,right_pair.ht)+1 #+1 since including the own node as well
    own_isBal = left_pair.isBal and right_pair.isBal and abs(left_pair.ht-right_pair.ht)<=1 #all True conditions    
    my_pair = BalPair(own_ht, own_isBal)

    return my_pair
if __name__ == '__main__':
#     arr = [50,25,12,None,None,37,30,None,None,None,75,62,None,70,None,None,87,None,None]
    arr = [50,25,12,10,None,None,None,None,6,None,7,None,9,None,None]
    root = construct(arr) 
    res_pair = isBalanced(root) 
    print('Is Balanced : ',res_pair.isBal)
```

## Largest BST Subtree


```python
class Node():
 
    def __init__(self, data,left,right):
        self.data = data
        self.left = left
        self.right = right
class Pair(): 
    def __init__(self, node,state):
        self.node = node
        self.state = state 

def construct(arr):
    stack = []
    root = Node(arr[0],None,None)
    stack.append(Pair(root,1))
    
    i = 0
    while len(stack) > 0:       
        #pick the top element Pair
        top = stack[-1]
        if top.state == 1:
            i+=1
            if arr[i] != None:
                left_node = Node(arr[i],None,None)
                top.node.left = left_node
                stack.append(Pair(left_node,1))
                top.state+=1
            else:
                top.node.left = None
                top.state+=1
                
        elif top.state == 2:
            i+=1
            if arr[i] != None:
                right_node = Node(arr[i],None,None)
                top.node.right = right_node
                stack.append(Pair(right_node,1))
                top.state+=1
            else:
                top.node.right = None
                top.state+=1
        else:
            stack.pop()
    return root

class BSTPair():
    def __init__(self,isBST,min_val,max_val,root,size):
        self.isBST = isBST
        self.min_val = min_val
        self.max_val = max_val
        # added features from isBalanced BST question
        self.root = root #largest_BST_root
        self.size = size #largest_BST_size
        #if a node is not BST, it might have a bigger subtree which is a BST, size here hold that subtree's size
        
def isBST(node):
    if node == None:
        return BSTPair(True,float('inf'),float('-inf'),None,0)
    #We are getting 3 info using Pair
    left_pair = isBST(node.left)
    right_pair = isBST(node.right)
    
    #tBST of left, tBST of right, node to be BST
    own_tBST = left_pair.isBST and right_pair.isBST and (node.data>left_pair.max_val and node.data<right_pair.min_val)
    #own min and own max
    own_min = min(node.data, min(left_pair.min_val , right_pair.min_val))
    own_max = max(node.data, max(left_pair.max_val , right_pair.max_val))
    '''Till here we check whether BST or not'''
    
    #Calculation of largest BST Node and it's size
    if own_tBST:
        own_root = node
        own_size = left_pair.size + right_pair.size + 1 #Size formula
    elif left_pair.size > right_pair.size:
        own_root = left_pair.root
        own_size = left_pair.size
        
    else:
        own_root = right_pair.root
        own_size = right_pair.size
        
    my_pair = BSTPair(own_tBST,own_min,own_max,own_root,own_size)
    
    return my_pair

if __name__ == '__main__':
    arr = [50,25,12,None,None,37,30,None,None,40,None,None,75,62,60,None,None,77,None,None,87,None,None]
    root = construct(arr) 
    own_res = isBST(root) 
    print('Largest BST Subtree is {} and size is {} '.format(own_res.root.data,own_res.size))
```

## Kth Largest in Binary Tree

```python
class Node():
 
    def __init__(self, data,left,right):
        self.data = data
        self.left = left
        self.right = right
class Pair(): 
    def __init__(self, node,state):
        self.node = node
        self.state = state 

def construct(arr):
    stack = []
    root = Node(arr[0],None,None)
    stack.append(Pair(root,1))
    
    i = 0
    while len(stack) > 0:       
        #pick the top element Pair
        top = stack[-1]
        if top.state == 1:
            i+=1
            if arr[i] != None:
                left_node = Node(arr[i],None,None)
                top.node.left = left_node
                stack.append(Pair(left_node,1))
                top.state+=1
            else:
                top.node.left = None
                top.state+=1
                
        elif top.state == 2:
            i+=1
            if arr[i] != None:
                right_node = Node(arr[i],None,None)
                top.node.right = right_node
                stack.append(Pair(right_node,1))
                top.state+=1
            else:
                top.node.right = None
                top.state+=1
        else:
            stack.pop()
    return root

def ceilAndFloor(node,data):
    if node == None:
        return
    #These global parameters are in heap , they dont depend on Euler path
    #The parameter like depth is in Stack, they change with Euler path(incr or decr wrt Euler path)
    global ceil
    global floor
    
    #Node se bada , lekin unn baro mein sabse chhota = ceil
    if node.data > data:
        if node.data<ceil:
            ceil = node.data
     
    #Node se chhota , lekin unn chhoto mein sabse bada = floor
    if node.data < data:
        if node.data>floor:
            floor = node.data
       
    #This code is to traverse the entire tree
    ceilAndFloor(node.left,data)
    ceilAndFloor(node.right,data)
    
def kthLargest(node,k):
    if node == None:
        return

    global factor
    global floor
    
    for i in range(k):
        ceilAndFloor(node,factor)
        factor = floor
        floor = float('-inf')#reinitialise for comparisons
    return factor

if __name__ == '__main__':
    arr = [50,25,12,None,None,37,30,None,None,None,75,62,None,70,None,None,87,None,None]
    root = construct(arr) 
    #Assign values here and in multisolver mention them as global var
    ceil = float('inf') #smallest among larger
    floor = float('-inf') #largest among smaller
    factor = float('inf')
    
    res = kthLargest(root,3)
    print('Kth Largest is : {}'.format(res))
```

## Width of Binary Tree

```python
class Node():
 
    def __init__(self, data,left,right):
        self.data = data
        self.left = left
        self.right = right
class Pair(): 
    def __init__(self, node,state):
        self.node = node
        self.state = state 

def construct(arr):
    stack = []
    root = Node(arr[0],None,None)
    stack.append(Pair(root,1))
    
    i = 0
    while len(stack) > 0:       
        #pick the top element Pair
        top = stack[-1]
        if top.state == 1:
            i+=1
            if arr[i] != None:
                left_node = Node(arr[i],None,None)
                top.node.left = left_node
                stack.append(Pair(left_node,1))
                top.state+=1
            else:
                top.node.left = None
                top.state+=1
                
        elif top.state == 2:
            i+=1
            if arr[i] != None:
                right_node = Node(arr[i],None,None)
                top.node.right = right_node
                stack.append(Pair(right_node,1))
                top.state+=1
            else:
                top.node.right = None
                top.state+=1
        else:
            stack.pop()
    return root

# ans[minHL,maxHL] , HL = Horizontal Level
def widthHelper(node,hl,ans):
    if node == None:
        return
    ans[0] = min(ans[0],hl)
    ans[1] = max(ans[1],hl)
    
    widthHelper(node.left,hl-1,ans)
    widthHelper(node.right,hl+1,ans)
    
def width(node):
    ans = [0]*2
    widthHelper(node,0,ans)
    return ans[1] - ans[0] + 1 #Total = b-a+1

if __name__ == '__main__':
    arr = [50,25,12,None,None,37,30,None,None,None,75,62,None,70,None,None,87,None,None]
    root = construct(arr) 
  
    
    res = width(root)
    print('Width is : {}'.format(res))
```

# Vertical Order of Binary Tree

```python
class Node():
 
    def __init__(self, data,left,right):
        self.data = data
        self.left = left
        self.right = right
class Pair(): 
    def __init__(self, node,state):
        self.node = node
        self.state = state 

def construct(arr):
    stack = []
    root = Node(arr[0],None,None)
    stack.append(Pair(root,1))
    
    i = 0
    while len(stack) > 0:       
        #pick the top element Pair
        top = stack[-1]
        if top.state == 1:
            i+=1
            if arr[i] != None:
                left_node = Node(arr[i],None,None)
                top.node.left = left_node
                stack.append(Pair(left_node,1))
                top.state+=1
            else:
                top.node.left = None
                top.state+=1
                
        elif top.state == 2:
            i+=1
            if arr[i] != None:
                right_node = Node(arr[i],None,None)
                top.node.right = right_node
                stack.append(Pair(right_node,1))
                top.state+=1
            else:
                top.node.right = None
                top.state+=1
        else:
            stack.pop()
    return root

# ans[minHL,maxHL] , HL = Horizontal Level
def widthHelper(node,hl,ans):
    if node == None:
        return
    ans[0] = min(ans[0],hl)
    ans[1] = max(ans[1],hl)
    
    widthHelper(node.left,hl-1,ans)
    widthHelper(node.right,hl+1,ans)
    


class vPair():
    def __init__(self,node,hl):
        self.node = node
        self.hl = hl
def verticalOrder(node):
    
    if node == None:
        return [[]]
    
    #Find width
    minMax = [0]*2
    widthHelper(node,0,minMax)
    total_width = minMax[1] - minMax[0] + 1 #Total = b-a+1
    
    res = [ [] for _ in range(total_width) ]
    
    mq = []
    mq.append(vPair(node,abs(minMax[0])))
    while(len(mq) > 0):
        ct = len(mq)
        for i in range(ct): 
            '''Remove'''
            removed_pair = mq.pop(0) 
            
            '''Work'''
            removed_node = removed_pair.node #get the node from pair
            removed_hl = removed_pair.hl #get the hl level from pair
            #Put it to the list of lists from your queue
            res[removed_hl].append(removed_node.data)
            
            '''Add child to queue'''
            if removed_node.left != None:
                mq.append(vPair(removed_node.left,removed_hl-1)) #Add the child
                
            if removed_node.right != None:
                mq.append(vPair(removed_node.right,removed_hl+1))
            
        
    return res

if __name__ == '__main__':
    arr = [50,25,12,None,None,37,30,None,None,None,75,62,None,70,None,None,87,None,None]
    #arr = [1,2,4,None,None,5,None,None,3,6,None,None,7,None,None]
    root = construct(arr) 
  
    
    ans = verticalOrder(root)
    print(ans)
    idx = 0
    for i in ans:
        print(str(idx) + " -> ",end = " ")
        idx+=1
        for j in i:
            print(str(j) + " ",end = " ");
        print()
```
